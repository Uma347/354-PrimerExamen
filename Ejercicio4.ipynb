Del anterior problema en PYTHON y WEKA realice un árbol de decisión basado en entropía, compare los resultados, esto debe repetirse al menos 10 splits y comparar las matrices de confusión y obtener la confiabilidad. Se recomienda comparar media y mediana de los resultados.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import tree
```

Realizando el arbol de desioion

```python
clasificador= tree.DecisionTreeClassifier()
data= pd.read_csv("diabetes.csv", sep=",", header=[0])
X = data.iloc[:, :-1].values
Y = data.iloc[:, 9].values 
from sklearn.preprocessing import LabelEncoder
label_encoder_y = LabelEncoder()
Y = label_encoder_y.fit_transform(Y)
clasificacion=clasificador.fit(X, Y)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
cv = StratifiedShuffleSplit(n_splits=10, test_size=0.50, random_state=0)
iteraciones = cv.split(X,Y)
from sklearn.tree import DecisionTreeClassifier
clasificador = DecisionTreeClassifier()
matrices=[[[]],[[]],[[]],[[]],[[]],[[]],[[]],[[]],[[]],[[]]]
confiabilidad=np.zeros(10)
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
for indice, (train_indice, test_indice) in enumerate(iteraciones):
    print("-----")
    print(indice)
    clasificador.fit(X[train_indice],Y[train_indice])
    y_obtenido = clasificador.predict(X[test_indice])
    print(Y[test_indice])
    print(y_obtenido)
    print("")
    print("Matriz de Confusión")
    print(confusion_matrix(y_obtenido,Y[test_indice]))
    print("Precisión")
    print(precision_score(y_obtenido,Y[test_indice]),"%")
    matrices[indice]=confusion_matrix(y_obtenido,Y[test_indice])
    confiabilidad[indice]=precision_score(y_obtenido,Y[test_indice])
    
```

Comparando resultados:

```python
for i in range(10):
    print("Split: ",i)
    print("Matriz de Confusión")
    print(matrices[i])
    print("Precisión")
    print(confiabilidad[i],"%")
    print("")
    
```

Media y mediana de los resultados:

```python
print("Media de Matrices de confusion  :")
print(np.mean(matrices,axis=0))
print("Mediana de Matrices de confusion:")
print(np.median(matrices,axis=0))
print()
print("Media de Presicion  :",np.mean(confiabilidad))
print("Mediana de Presicion:",np.median(confiabilidad))
```
